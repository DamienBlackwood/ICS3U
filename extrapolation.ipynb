```python
# Sir if it says no file found or something then just plug your file in cus its not updating through cloud anymore for me and its very annoying, thank you

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
uploaded = files.upload()

df = pd.read_csv("population_data.csv")

print("Chose one")
print("1. Estimate the population over 20 years")
print("2. Display a bar graph of the population trend.")
print("3. Display a line graph of the population trend.")
choice = int(input("Pick any, (1, 2, 3): "))

if choice == 1:
    # so here im basically assigning the two giving values, years and population to the assigened variables so I can calcluate using slope formula.
    x = df['Year'].values.reshape(-1, 1)
    y = df['Population'].values.reshape(-1, 1)

    # the reason I used the reshape thing is because it helps me to find the linear regression like for the extrapolation so it is necessary it makes the 1D array turn into 2D1
    xmean = np.mean(x)
    ymean = np.mean(y)
    h = np.sum((x - xmean) * (y - ymean))
    y = np.sum((x - xmean)**2)
    slope = h / y
    line = ymean - slope * xmean

    newyr = 2100 + 20
    newpop = slope * newyr + line

    # so im just plotting the trend
    plt.plot(df['Year'], df['Population'], label="Population Trend")
    plt.plot([newyr], [newpop], "ro--", label="Extrapolation")
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.title('Population Trend Over The Years')
    plt.legend()
    plt.grid(True)
    plt.show()

    # this is the populatin in 20 years
    print(f"The estimated population in {newyr} is {newpop:0f}.")


# this part is basic stuff that just prints out what is shown in the CSV

elif choice == 2:
  plt.bar(df['Year'], df['Population'])
  plt.xlabel('Year')
  plt.ylabel('Population')
  plt.title('Population Trend Over The Years')
  plt.show()

elif choice == 3:
  plt.plot(df['Year'], df['Population'])
  plt.xlabel('Year')
  plt.ylabel('Population')
  plt.title('Population Trend Over The Years')
  plt.show()

```
