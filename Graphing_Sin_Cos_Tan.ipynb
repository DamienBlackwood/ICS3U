```python
import matplotlib.pyplot as plt
import numpy as np

print("""
1. Sin
2. Cos
3. Tan
4. Special Equation
""")

# Give me the input for y = sin (your number)x

# THESE ARE THE FUCNTIONS FOR THE TYPE OF EQUATIONS #

def sin():
  s = float(input("Enter the coefficient for your sin equation (like y = sin(Your Number * x))"))

  x = np.linspace(0 , 2 * np.pi, 2000)
  y = np.sin(s*x)

  plt.figure(figsize=(10,2.5))
  plt.plot(x, y, label=f'y = sin({s} * x)')
  plt.ylim(-1, 1)
  plt.grid(True)
  plt.title(f'y = sin({s} * x)')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.show()

def cos():
  c = int(input("What is your number? "))

  x = np.linspace(0, 2 * np.pi, 2000)
  y = np.cos(c*x)


  plt.figure(figsize=(10,2.5))
  plt.plot(x, y, label=f'y = sin({c} * x)')
  plt.ylim(-1, 1)
  plt.grid(True)
  plt.title(f'y = cos({c} * x)')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.show()

def tan():
  t = int(input("Enter the coefficient for y = tan(t * x)"))

  x = np.linspace(-2 * np.pi, 2 * np.pi, 10000)
  y = np.tan(t * x)

  y = np.ma.masked_array(y, np.abs(y) > 10, y)

  plt.figure(figsize=(10,5))
  plt.plot(x, y, label=f'y = tan({t} * x)')

  plt.ylim(-6.5, 6.5)
  plt.title(f'y = tan({t} * x)')
  plt.xlabel('x')
  plt.ylabel('y')
  plt.grid(True)
  plt.show()




def scrape():
    equation = input("Enter an equation (e.g., 'sin 3x + 5' or '4 * sin 2x'): ")


    x = np.linspace(-2 * np.pi, 2 * np.pi, 10000)
    y = np.zeros_like(x)

    # now im just like deletintg the spaces or whatever the technical term is
    equation = equation.replace(' ', '')
    equation = equation.replace('sin', 'sin(').replace('cos', 'cos(').replace('tan', 'tan(')
    equation = equation.replace('x)', 'x)').replace(')sin', ') * sin').replace(')cos', ') * cos').replace(')tan', ') * tan')

    # so if its negative or positive it'll blasically split it
    terms = equation.replace('-', '+-').split('+')

    # now this is just the NOT simple part where im scraping or parsing or whatever you wanna call it to get all the coefs, and x's an then add it to the y total equation as you can see in the last line
    try:
        for term in terms:
            if 'sin(' in term:
                # Handle sin function
                coef_str = term.split('sin(')[1].strip(')')
                coef = float(coef_str.split('x')[0]) if 'x' in coef_str else 1
                # Add the sin term to y
                y += np.sin(coef * x)
            elif 'cos(' in term:
                # Handle cos function
                coef_str = term.split('cos(')[1].strip(')')
                coef = float(coef_str.split('x')[0]) if 'x' in coef_str else 1
                y += np.cos(coef * x)
            elif 'tan(' in term:
                # Handle tan function
                coef_str = term.split('tan(')[1].strip(')')
                coef = float(coef_str.split('x')[0]) if 'x' in coef_str else 1
                y += np.tan(coef * x)
                # Mask large y values to handle tangent spikes
                y = np.ma.masked_where(np.abs(y) > 10, y)
            # OKAY SO THIS TOOK ME FOREVER TO GET RID OF THOSE GOD DAMN LINES, so the abs thing just returns the number regardless of whatever the sign is and I NEED THAT so i can make sure that the lines are masked away, so by the absolute values it checks if they are greater than 10, then it just cacluates it with the mask

            else:
                # Handle linear terms or constants
                if 'x' in term:
                    coef, variable = term.split('x')
                    coef = float(coef) if coef else 1
                    y += coef * x
                else:
                    # Handle constant terms
                    y += float(term)

    except ValueError as e:
        print("Invalid input. Please check your equation format.")

    # Plot the resulting function
    plt.figure(figsize=(12, 6))
    plt.plot(x, y)
    plt.title(f'y = {equation}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)

    # Set y-limits based on the function type for clarity
    if 'sin(' in equation or 'cos(' in equation:
        plt.ylim(-1, 1)  # Set y limits for sin and cos functions
    elif 'tan(' in equation:
        plt.ylim(-6.5, 6.5)  # Set y limits for tan function

    plt.show()

opt = int(input("What's type of graph do you want to see? "))

if opt == 1:
  sin()
elif opt == 2:
  cos()
elif opt == 3:
  tan()
elif opt == 4:
  scrape()
else:
  print("What the heck did you put")
```
